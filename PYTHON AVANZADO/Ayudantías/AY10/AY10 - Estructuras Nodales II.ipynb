# Representacion con nodos

El approach naive es tomar los datos duros, que estÃ¡n en una plantilla excel, y hacer fuerza bruta para ver si existe un camino. Eso es muy costoso, nos demorarÃ­amos aproximadamente $2\cdot 10^{115}$ aÃ±os considerando 100000 operaciones por segundo. Entonces necesitamos una forma mÃ¡s fÃ¡cil de calcular estos caminos. AquÃ­ entran los grafos.

Los grafos al igual que las listas ligadas y arboles (que realmente son casos especiales de grafos) pueden ser representados por las estructuras llamadas nodos. El atributo ```self.vecinos``` es una lista que representa los nodos vecinos a los que se puede acceder desde un determinado nodo.

````python
class Nodo:
  def __init__(self, valor):
    self.valor = valor
    self.vecinos = []
    
  def __repr__(self):
    return str(self.valor)
````

Para el grafo, nos conviene tener referencia a todos los nodos y ademas que sean fÃ¡cil de acceder. Por eso utilizaremos un diccionario donde la ```key``` sera el valor del grafo (en algunos casos se utilizan IDs) y el ```value``` sera el nodo asociado a ese valor.

````python
# Clase grafo
class Grafo:
  def __init__(self):
    self.nodos = {}
````

# Operaciones

Por simplicidad utilizaremos un grafo no dirigido

````python
# Agregar Nodo
class Grafo(Grafo):
    def agregar_nodo(self, valor):
        if valor not in self.nodos:
            self.nodos[valor] = Nodo(valor)
````

````python
# Conectar nodos
class Grafo(Grafo):
    def conectar_nodos(self, valor1, valor2):
        if self.nodos[valor2] not in self.nodos[valor1].vecinos:
            self.nodos[valor1].vecinos.append(self.nodos[valor2])
    
    # como es No Dirigido tambien debemos conectar al vecino con el nodo
        if self.nodos[valor1] not in self.nodos[valor2].vecinos: 
            self.nodos[valor2].vecinos.append(self.nodos[valor1])
````

````python
# Desconectar nodos
class Grafo(Grafo):
    def desconectar_nodos(self, valor1, valor2):
        self.nodos[valor1].vecinos.remove(self.nodos[valor2])
        self.nodos[valor2].vecinos.remove(self.nodos[valor1]) # tambien desconectamos al vecino
````

````python
# Quitar nodo
class Grafo(Grafo):
    def quitar_nodo(self, valor):
        if valor in self.nodos:
            nodo = self.nodos[valor]
            for otro_nodo in nodo.vecinos:
                otro_nodo.vecinos.remove(nodo)
            del self.nodos[valor]
````


Finalmente podemos cargar los datos del problema creando el grafo

````python
def cargar_grafo():
    grafo = Grafo()
    for nodo in nodos.strip().split('\n'):
        grafo.agregar_nodo(nodo.strip())
    for relacion in relaciones.strip().split('\n'):
        n1, n2 = relacion.strip().split(' ')
        grafo.conectar_nodos(n1, n2)
    return grafo
````

# Algoritmos de Busqueda
Al igual que con los arboles podemos recorrer los grafos con los siguientes algoritmos:

### BFS: Breadth-first search
El algoritmo BFS recorre exhaustivamente el grafo dado un nodo de partida. Si un nodo no fue visitado en el recorrido, significa que **no es alcanzable** desde ese punto de partida. Se utiliza una **cola** para guardar los nodos a visitar.

BFS recorre los vecinos cercanos a una distancia de uno del punto de partida, despues a una distancia de 2 y asÃ­ progresivamente.

Es especialmente util para encontrar el camino mas corto entre 2 nodos.

<img src="imagenes/bfs_graph.gif">

### DFS: Depth-first search
Al igual que BFS, el algoritmo DFS recorre exhaustivamente el grafo, dado un punto de partida. Si un nodo no fue visitado en el recorrido, significa que **no es alcanzable** desde ese punto de partida. Se utiliza un **stack** para guardar los nodos a visitar.

DFS trata de llegar lo mas lejos posible a partir del nodo inicial, por eso es una busqueda en profundidad.

DFS tambien encuentra el camino entre 2 nodos, pero a diferencia de BFS este no siempre es el camino mas corto

# Busqueda sin Visitados

**LA GRAN DIFERENCIA** respecto al recorrido con arboles, es que en los nodos se pueden formar ciclos en su recorrido. Po lo tanto si no guardamos los nodos que ya hemos visitados podemos quedar atrapados en este ciclo. Una forma de **NO IMPLEMENTAR** BFS es el siguiente:

````python
from collections import deque

def bfs_malo(grafo, inicio, final):
    queue = deque([inicio])

    while len(queue) > 0:
        valor = queue.popleft()
        print(valor)
        vertice = grafo.nodos[valor]
        # si lo encontramos retornamos True
        if valor == final:
            return True
        for vecino in vertice.vecinos:
            queue.append(vecino.valor)
    return False
````


Intentaremos recorrer el grafo para saber si es que existen la conexion. **Spoiler** sale mal :(

````python
grafo = cargar_grafo()

bfs_malo(grafo, 'capibara', 'castor')
````

# Busqueda con Visitados

A continuacion implementaremos un algoritmo dfs recursivo con la diferencia que si guarda los nodos visitados en un **set** para evitar volver a recorrerlos. Como es recursivo este set se pasa a cada llamada recursiva que realiza el algoritmo.

````python
def dfs(grafo, inicio, final, visitados):
    print(inicio)
    if inicio == final:
        return True
    visitados.add(inicio)
    nodo = grafo.nodos[inicio]
    for vecino in nodo.vecinos:
        if vecino.valor not in visitados:
            resultado = dfs(grafo, vecino.valor, final, visitados)
            if resultado:
                return resultado
    return False
````

# Armar Camino

En muchas ocaciones es util armar un camino que muestre los nodos en el orden que fueron recorridos, para esto debemos:

1. Crear un diccionario de padres que tendra la ```key``` el nodo Hijo y el ```value``` el nodo Padre
2. Definimos el valor del nodo inicio en ```None```
3. Cada vez que agregamos un nodo a visitar en la cola o el stack debemos darle la referencia al padre en el diccionario

Ademas si estamos utilizando un algoritmo recursivo debemos pasar este diccionario en cada llamada recursiva que hacemos

Cuando ya tengamos el diccionario de padres utilizaremos la siguiente funcion para para armar el camino:

````python
def camino(diccionario_padres, final):
    # recibe el diccionario de padres y el valor (o nodo) final
    camino = []
    camino.append(final)
    while diccionario_padres[final] is not None:
        camino.append(diccionario_padres[final])
        final = diccionario_padres[final]
    camino.reverse()
    return camino
````

Implementaremos un bfs que retornara el diccionario de padres y el nodo final para poder utilizarlos en la funcion anterior.

````python
from collections import deque

def bfs_con_camino(grafo, inicio, final):
    visitados = set()
    por_visitar = deque([(inicio)])
    
    # (1) creamos el diccionario
    padres = {}
    # (2) fijamos el padre del nodo inicial en None
    padres[inicio] = None
    while por_visitar: 
        valor = por_visitar.popleft()
        if valor in visitados:
            continue
        if valor == final:
            break
    
        visitados.add(valor)
        nodo = grafo.nodos[valor]
    
        for vecino in nodo.vecinos:
            if vecino.valor not in visitados:
                por_visitar.append(vecino.valor)
                # (3) Se le da un padre al nodo agregado
                padres[vecino.valor] = valor

    return padres, final
````

````python
grafo = cargar_grafo()

padres, final = bfs_con_camino(grafo, 'capibara', 'cebra')

camino(padres, final)
````
